Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a=float(input("nhập chiều a:"))
b=float(input("nhập chiều b:"))
c=float(input("nhập chiều c:"))
if a+b>c and a+c>b and b+c>a and a>0 and b>0 and c>0:
    print("3 cạnh trên có thể tạo thành 1 tam giác")
    if a==b and b==c:
        print("3 cạnh trên có thể tạo thành 1 tam giác đều")
    elif a==b or b==c or a==c:
        print("3 cạnh trên có thể tạo thành 1 tam giác cân")
    elif a**2==b**2+c**2 or b**2==a**2+c**2 or c**2==a**2+b**2:
        print("3 cạnh trên có thể tạo thành 1 tam giác vuông")
    else :
        print("3 cạnh trên có thể tạo thành 1 tam giác thường")
    p=(a+b+c)/2
    S=(p*(p-a)*(p-b)*(p-c))**0.5
    print ("diện tích tam giác là:", S)
    

else:
    print("3 cạnh trên không thể tạo thành 1 tam giác")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python

a, b, c = map(float, input("nhập 3 số thực:").split())
delta = b**2 - 4*a*c
if delta < 0:
    print("phương trình vô nghiệm")
elif delta == 0:
    x = -b/(2*a)
    print(f"phương trình có nghiệm kép x = {x}")
else:
    x1 = (-b + delta**0.5)/(2*a)
    x2 = (-b - delta**0.5)/(2*a)
    print(f"phương trình có 2 nghiệm phân biệt x1 = {x1} và x2 = {x2}")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a = int(input("hoành độ tâm:"))
b = int(input("tung độ tâm:"))
c = int(input("bán kính:"))
d = int(input("hoành độ điểm:"))
e = int(input("tung độ điểm:"))
s = ((d - a)**2 + (e - b)**2)**0.5
if s < c:
    print("điểm nằm trong đường tròn")
elif s == c:
    print("điểm nằm trên đường tròn")       
else:
    print("điểm nằm ngoài đường tròn")
```
