Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.


```python
def so(a):
    d ={}
    for i in a : 
        if i in d : 
            d[i]  += 1 
        else : 
            d[i]= 1 
    s = 0 
    for i in d : 
        if d[i]<i:
            s += d[i]
        else : 
            s += d[i]-i
    return s 
a = [1, 2, 2, 3, 3, 3, 3]
print(so(a))
         
```

dần, lẻ giảm dần; ; 

```python
def so(a):
    moi =[]
    cu = []
    for i in a : 
        if i % 2 == 0 : 
            moi.append(i)
        else : 
            cu.append(i)
    moi.sort()
    cu.sort(reverse = True)
    return moi + cu 
a = [5, 2, 8, 3, 1, 4]
print(so(a))
        
```

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập
vào từ bàn phím 

```python
a = list(map(int, input().split()))

n = len(a)

dp = [1] * n

for i in range(n):
    for j in range(i):
        if a[j] < a[i]:
            dp[i] = max(dp[i], dp[j] + 1)

print(max(dp))
```
