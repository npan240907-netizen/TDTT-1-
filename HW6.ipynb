Nhâ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải


```python
n, m = map(int,input().split())
a = [list(map(int, input().split())) for _ in range(n)] 
for i in a : 
    for x in i : 
        print(f"{x:>4}", end="") 
    print()
```

Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
n = int(input())
a = [list(map(int,input().split())) for i in range(n)]
print("đường chéo chính:")
for i in range(n) : 
    print(a[i][i],end=" ")
print("\nđường chéo phụ:")
for i in range(n):
    print(a[i][n-1-i],end=" ")
```

Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k)

trong list

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1


```python
def so(a, k):
    for i in range(len(a)): 
        if a[i]== k : 
            return i 
    return -1
a = list(map(int,input().split()))
k = int(input())
print(so(a, k))
```

Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
s = input()
d= {}
for i in range(len(s)):
    if s[i]not in d :
        d[s[i]]=(i,)
    else : 
        d[s[i]]+=(i,)
print(d)
```

VIết hàm xóa đi các số <x có trong mảng

```python
def xoa(a, x):
    kq = []
    for i in a : 
        if i >= x: 
            kq.append(i)
    return kq
a = list(map(int,input().split()))
x = int(input())
print(xoa(a, x))
```

Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def so(a):
    vi_tri = 0 
    for i in range(len(a)):
        if a[i] == 0:
            a[i],a[vi_tri] = a[vi_tri], a[i]
            vi_tri += 1
    return a
a = list(map(int,input().split()))
print(so(a))
```

Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong

Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

```python
def tong(a):
    t = 0 
    for i in range(1,len(a)-1):
        if a[i]>a[i-1] and a[i]> a[i+1]:
            t += a[i]
    return t 
a = list(map(int,input().split()))
print(tong(a))
```

Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1


```python
def so(a, x):
    d = []
    for i in range(len(a)):
        if a[i]< x : 
            d.append(i)
    return d 
a = list(map(int,input().split()))
x = int(input())
print(so(a, x))
```

Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
def so(a):
    tang = True 
    giam = True 
    for i in range(1 ,len(a)):
        if a[i] <= a[i-1]:
            tang = False 
       
        if a[i] >= a[i-1]:
            giam = False 
    if tang : 
        return "Tăng thực sự"
    if giam : 
        return "Giảm thực sự "
    else : 
        return "vô danh "
a = list(map(int,input().split()))
print(so(a))
```
