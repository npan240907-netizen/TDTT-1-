Viết hàm tìm kiếm nhị phân trên một danh sách đãđược sắp xếp tăng dần.Trả về chỉ số của phần tửhoặc -1 nếu không tìmthấy. mẫu hàm: defbinary_search(arr, left,right, target):


```python
def binary_search(arr,left,right, target):
    if left > right : 
        return -1 
    mid = (left + right)//2
    if arr[mid] == target : 
        return mid
    elif arr[mid] > target:
        return binary_search(arr,left,mid -1, target)
    else : 
        return binary_search(arr,mid+1,right, target)
a = [1, 3, 5, 7, 9]
print(binary_search(a, 0, len(a)-1, 7))
            
```

Cho một danh sách các số
nguyên nums, hãy tìm độ
dài của dãy con tăng
nghiêm ngặt dài nhất
(không nhất thiết phải liền kề).

```python
def so(n):
    x = [1]*len(n)
    for i in range(len(n)):
        for j in range(i):
            if n[j]<n[i]:
                x[i]= max(x[i],x[j]+1)
    return max(x)
n = [10, 9, 2, 5, 3, 7, 101, 18]
print(so(n))
```
