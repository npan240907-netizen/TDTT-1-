Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x):
    s = str(x)
    n = len(s)
    x= 0 
    for i in s:
        x += int(i)**n
    return x == int(s)
print(is_armstrong(153))
```

nguyên x

```python
def so (x):
    if x < 0:
        return False
    i = int(x**0.5) 
    return i * i == x
print(so(5))
```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return("int")
    except:
        try:
            float(s)
            return("float")
        except:
            return("neither")
print(detect_number_type("123"))
```

Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a, b, c):
    return a==b*c or b==a*c or c==a*b
print(is_one_product_of_others(2, 4, 8))
```

Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.

 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a, b, c):
    s = b**2-4*a*c
    if s > 0 :
        return 2 
    elif s == 0: 
        return 1 
    else :
        return 0
print(quadratic_roots_count(1, -3, 2))
```

Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    return (y2-y1)*(x3-x1) == (y3-y1)*(x2-x1)
print(are_collinear(1, 2, 2, 3, 3, 4))
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    x1, y1 = quad[0]
    x2, y2 = quad[1]
    x3, y3 = quad[2]
    x4, y4 = quad[3]
    xp, yp = p
    area_quad = 0.5 * abs(x1*y2 + x2*y3 + x3*y4 + x4*y1 - y1*x2 - y2*x3 - y3*x4 - y4*x1)
    area1 = 0.5 * abs(xp*y2 + x2*y3 + x3*yp - yp*x2 - y2*x3 - y3*xp)
    area2 = 0.5 * abs(x1*yp + xp*y3 + x3*y1 - yp*x1 - y3*x1 - x3*yp)
    area3 = 0.5 * abs(x1*y2 + x2*yp + xp*y1 - y2*x1 - yp*x2 - x1*yp)
    area4 = 0.5 * abs(xp*y2 + x2*y4 + x4*yp - yp*x2 - y4*xp - x4*yp)
    return area_quad == area1 + area2 + area3 + area4
print(point_in_quadrilateral(quad, p))   
```

Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

Viết hàm tính hệ số nhị thức C(n, k).


```python
def hệ (n):
    s = 1 
    for i in range(1, n + 1):
        s *= i
    return s
def C(n, k):
    return hệ(n) // (hệ(k) * hệ(n - k))
print(C(5, 2))
```

Viết hàm tính tổng bình phương các số từ 1 đến n.
